## functions of $r, s, \psi, l, b$

$$ \cos\psi = \cos b \cos l $$
$$ \psi = \arccos (\cos b \cos l) $$
where $b$ is in the standard convention, ranging from -90$^\circ$ to 90$^\circ$
(galprop has $b$ in its own convention and needs to be translated)

```python
def b_val(b):
    """
    returns: [degrees] new b in standard convention
    converts GALPROP value for b (-179.75 to -0.25) to the standard convention of (-90 to 90)
    central value becomes 0.25, so new b ranges from (-89.75 to 89.75)
    """
    return b + 90


def Psi(b, l):
    """
    returns: [radians] value of angle psi
    ********
    b: [degrees] galactic latitude in GALPROP convention
    l: [degrees] galactic longitude in GALPROP/standard convention
    """
    # first convert b to standard convention using b_val(b)
    # then convert b and l to radians
    return np.arccos( np.cos(np.radians(b_val(b))) * np.cos(np.radians(l+90)) )
    
```

Radial coordinate $r$ is related to line of sight variable $s$ by
$$ r = \sqrt{R_\odot^2 + s^2 - 2 R_\odot s \cos\psi} $$

Maximum value of line of sight variable $s$ for a given $(b, l)$ is determined by the size of the halo (i.e. the max $r$ value) and $R_\odot$:
\begin{align}
s_\text{max} &= \sqrt{r^2 - R_\odot^2 \sin^2 \psi} + R_\odot\cos\psi \Bigg|_{r = R_\text{halo}} \\
&= \sqrt{R_\text{halo}^2 - R_\odot^2 \sin^2 \psi} + R_\odot\cos\psi
\end{align}

```python
def R(s, b, l):
    """
    returns: [kpc] radius, as a function of s, b, l
    ********
    s: [kpc] line of sight coordinate
    b: [degrees] galactic latitude in GALPROP convention
    l: [degrees] galactic longitude in GALPROP/standard convention
    R_odot = 8.0 # [kpc] Sun's distance from centre of MW
    """
    psi = Psi(b, l) # [radians]
    return np.sqrt(R_odot**2 + s**2 - 2 * R_odot * s * np.cos(psi)) # [kpc]


def s_max(b, l, R_halo):
    """
    returns: [kpc] maximum value of line-of-sight coordinate, as a function of angles b, l, and the halo scale R_halo
    ********
    b: galactic latitude [degrees] in GALPROP convention
    l: galactic longitude [degrees] in GALPROP/standard convention
    R_halo: MW halo radius [kpc]
    R_odot = 8.0 # Sun's distance from centre of MW [kpc]
    """
    psi = Psi(b, l) # [radians]
    return (np.sqrt(R_halo**2 - R_odot**2 * (np.sin(psi))**2) + R_odot * np.cos(psi)) # [kpc]
    
```

# $v_\text{esc}(r)$

$$ v_\text{esc}(r) = \sqrt{\frac{2 G M}{r}} $$
where $M$ is the mass enclosed at radius $r$ measured from the galactic centre.

$M$ can be calculated by volume integral of total density (baryon and DM):

\begin{align}
M &= \int \rho_\text{tot} ~d\tau \\
&= \int [ \rho_\text{b}(r, b, l) + \rho_\text{DM}(r) ] ~d\tau \\
&= \int_0^{2\pi} d\phi \int_0^\pi d\theta \sin\theta \int_0^r dr'~ r'^2 [ \rho_\text{b}(r', b, l) + \rho_\text{DM}(r') ] \\
&= \int_0^{2\pi} d\phi \int_0^\pi d\theta \sin\theta \int_0^r dr'~ r'^2 \rho_\text{b}(r', b, l) + 4\pi \int_0^r dr'~ r'^2 \rho_\text{DM}(r')
\end{align}

```python
def density_DM(r):
    return rho_s / ( (r/r_s) * (1 + r/r_s)**2 ) # [GeV/cm^3]
```

```python
def density_interpolate(r, b, l):
    """
    returns: [kg/cm^3] density at a given radius, interpolated for (b, l) from gridded data
    """
    r_index = bisect.bisect(r_bins, r) - 1 # find radial bin index
    L = ls
    B = bs
    values_grid = hdul_CO[0].data[r_index] # reduces 3D grid (r, b, l) to 2D (b, l)
    interpolating_fx = RegularGridInterpolator((B,L), values_grid)
    
    return interpolating_fx(np.array([b, l]))


def M_MW(r_esc):
    """
    returns: [kg] mass of MW within radius r_esc [kpc]
    r_esc: [kpc] 
    *** CURRENTLY ONLY USING DM DENSITY ***
    to include baryons, need to express l, b in terms of galactocentric angles theta, phi
    """
    def f(r):
        """
        returns: [GeV/cm^3 * kpc^3] mass
        """
        return 2*np.pi * 2 * r**2 * rho_s / ( (r/r_s) * (1 + r/r_s)**2 ) # [GeV/cm^3] * kpc^3
    # convert to kg
    mass_kg = scipy.integrate.quad(f, 0, r_esc)[0] * cm_per_kpc**3 * 1/GeV_per_kg # [kg]
    mass_odot = mass_kg / M_sun # [solar masses]
    # print("{:2.2e} solar masses".format(mass_odot))
    # MW galaxy should be approximately 1.2e12 solar masses; currently off ~ 1 orders of magnitude
    return mass_kg
    

def v_esc(s, b, l):
    """
    returns: [km/s] escape velocity as function of radius
    *** CURRENTLY ONLY ASSUMES DM MAKING UP MASS ***
    """
    r = R(s, b, l) # [kpc]
    # print(r)
    M = M_MW(r) # [kg]
    # M = 1.2e12 * M_sun
    # convert r from kpc to cm to m
    # convert to km/s
    return 3 * np.sqrt(2 * G * M / (r * cm_per_kpc / 100)) / 1000 # [km/s]
```

# velocity dispersions $\sigma_i(r), \sigma_\chi(r)$

DM velocity dispersion approximately linear-log from https://arxiv.org/pdf/2111.03076.pdf; using automeris.io data extraction for $r$ < 10 kpc:
$$ \sigma_\text{3D} \approx -42.4 \log_{10}r + 382.2 $$

Baryon velocity dispersion assumed to be constant 10 km/s.


```python
def v_dispersion(particle, s, b, l): # to be updated with functions of r from Louis & Addy
    """
    returns: 3D velocity dispersion [km/s]
    **********
    particle: 'DM', 'He4', 'C12', 'N14', 'O16', 'Fe57'
    b: galactic latitude [degrees] in GALPROP convention
    l: galactic longitude [degrees] in GALPROP/standard convention
    """
    r = R(s, b, l) # [kpc] # finds radius for a given (s, b, l) in order to find the dispersion at those coordinates
    # if r > r_bins[-1]: # if radius is beyond what is provided by GALPROP, return error
    #     print('no baryon density for this radius, try reducing R_max')
    #     return
    # print(r)  
    if particle == 'DM':
        """
        dispersion function taken from 2111.03076 Figure 1: r in units kpc, velocity dispersion in units km/s
        """
        return (-42.4 * np.log10(r) + 382.2) # [km/s]
    # elif particle == 'He4': # copy for each potential target
        # return function of r
    elif particle in ['He4', 'C12', 'N14', 'O16', 'CO', 'Fe57']:
        return 10 # [km/s]
    else:
        raise ValueError("particle must be one of 'DM', 'He4', 'C12', 'N14', 'O16', 'Fe57'")
        
```

# density distributions $\rho_i(r), \rho_\chi(r)$

The distribution functions as written are normalized s.t.
\begin{equation}
  \rho_i (r) = \int d^3 v_i \, f_i(r,\vec{v}_i)
\end{equation}

\begin{equation}
  \rho_\chi (r) = \int d^3 v_\chi \, f_\chi(r,\vec{v}_\chi)
\end{equation}

## DM density distribution

Assuming an NFW DM density distribution
\begin{equation}
    \rho(r) = \frac{\rho_s}{\frac{r}{r_s} \left[ 1 + \left( \frac{r}{r_s} \right) \right]^{2}},
\end{equation}

## density distribution functions $\rho_i(r), \rho_\chi(r)$

probably want to change this so that if r>50 kpc (GALPROP limit), return density of zero

```python
def density(particle, s, b, l):
    """
    returns: [GeV/cm^3] mass density, at radius r
             (which is also the normalization factor for the velocity distribution)
    **********
    particle: 'DM', 'He4', 'C12', 'N14', 'O16', 'CO', 'Fe57'
    l: [kpc] line of sight distance
    b: [degrees] galactic latitude in GALPROP convention
    l: [degrees] galactic longitude in GALPROP/standard convention

    constants:
    rho_s = 0.51 # DM radial scale at (r = R_odot) for MW NFW profile [GeV/cm^3]; https://arxiv.org/abs/1906.08419

    conversion factors:
    GeV_per_kg = 5.625e26
    """
    r = R(s, b, l) # [kpc] # finds radius for a given (s, b, l) in order to find the dispersion at those coordinates
    #print(r)
    
    if particle == 'DM':
        return rho_s / ( (r/r_s) * (1 + r/r_s)**2 ) # [GeV/cm^3]
    # elif particle == 'He4': # copy for each potential target
        # return function of r
    elif particle in ['He4', 'C12', 'N14', 'O16', 'Fe57']:
        if r > r_bins[-1]: # if radius is beyond what is provided by GALPROP, return zero
            #print('no baryon density for this radius, try reducing R_max')
            return 0
        r_index = bisect.bisect(r_bins, r) - 1 # find radial bin index
        # print(r_index)
        # no interpolating, b and l need to be in the bs, ls arrays:
        # rho_i = hdul_CO[0].data[r_index, bs.index(b), ls.index(l)] # [kg/cm^3]
        # density interpolating function:
        rho_i = density_interpolate(r, b, l)[0] # [kg/cm^3]
        #print(rho_i)
        # convert from [kg/cm^3] to [GeV/cm^3]: 
        return rho_i * GeV_per_kg # [GeV/cm^3]
    else:
        raise ValueError("particle must be one of 'DM', 'He4', 'C12', 'N14', 'O16', 'Fe57'")
        
```

# J-factor tests

For decaying DM:
$$ J(\psi) = \int_\text{LOS} \frac{ds}{r_\odot} \left( \frac{\rho(r(s,\psi))}{\rho_\odot} \right) $$

<!-- $$ \mathcal{J}(\psi) = \int_\text{LOS} \rho(r(s,\psi))^2 ~ds $$ -->

since $\cos\psi = \cos b \cos l$, if we set $b \approx 0$,
$$ \cos\psi \approx \cos l$$

(using $l$ gives us the typical $\psi$ range from 0 to 180)

$$ \int d^3v f(v) = \frac{\rho_{CO}}{m_{CO}} $$

$$ f(v) = \tilde{f}(v) \frac{\rho_{CO}}{m_{CO}} $$

$$ \int d^3v \tilde{f}(v) = 1 $$

$$ \int ds \int d^3v_\chi f(v_\chi) \int d^3v_b f(v_b) ~\rho_\chi \rho_b \sim \int ds \rho_\chi^2 $$
^ compare this to J-factor calculation
to check baryon density

```python
def Jfactor_decay_psi(psi):
    # https://arxiv.org/pdf/1012.4515.pdf
    r_odot = R_odot # 8.33 # [kpc] pg 33
    rho_odot = rho_s # 0.3 # [GeV/cm^3] pg 33

    def f(s):
        r = np.sqrt(r_odot**2 + s**2 - 2*r_odot*s*np.cos(np.radians(psi)))
        integrand = (1/r_odot) *  ( rho_s / ( (r/r_s) * (1 + r/r_s)**2 ) ) / rho_odot
        
        # integrand = (1 / r_odot) * (density(particle, s, b, l) / rho_odot)**2
        return integrand

    return scipy.integrate.quad(f, 0, 10)[0] # [km^3 s^-3 MeV^-2]


def Jfactor_decay(b, l):
    # https://arxiv.org/pdf/1012.4515.pdf
    r_odot = R_odot # 8.33 # [kpc] pg 33
    rho_odot = rho_s # 0.3 # [GeV/cm^3] pg 33

    def f(s):
        r = R(s, b, l)
        # r = np.sqrt(r_odot**2 + s**2 - 2*r_odot*s*np.cos(np.radians(psi)))
        integrand = (1/r_odot) *  density('DM', s, b, l) / rho_odot
        
        # integrand = (1 / r_odot) * (density(particle, s, b, l) / rho_odot)**2
        return integrand

    return scipy.integrate.quad(f, 0, 10)[0] # [km^3 s^-3 MeV^-2]
```

# $v_\chi$ integral
\begin{align}
\int d^3 v_\chi \frac{f_\chi (\vec{v}_\chi, r)}{m_\chi} \sigma(v_\text{rel}) v_\text{rel} &\sim \int d^3 v_\chi \frac{1}{E_\chi} \exp \left[ -\frac{|\vec{v}_\chi|^2}{2\sigma_\chi(r)^2} \right] |\vec{v}_\chi|^2 \frac{\delta(v_\chi - V_\chi^+)}{V_\chi^+} \\
&= \int d^3 v_\chi \frac{1}{m_\chi + \frac{1}{2} m_\chi v_\chi^2} \exp \left[ -\frac{|\vec{v}_\chi|^2}{2\sigma_\chi(r)^2} \right] |\vec{v}_\chi|^2 \frac{\delta(v_\chi - V_\chi^+)}{V_\chi^+} \\
&= \frac{1}{m_\chi + \frac{1}{2} m_\chi (V_\chi^+)^2}\exp \left[ -\frac{(V_\chi^+)^2}{2\sigma_\chi(r)^2} \right] V_\chi^+
\end{align}

where

$$ V_\chi^+ (v_i) = \left[ \frac{2(\Delta E - m_\chi - \frac{1}{2}m_N v_i^2)}{m_\chi} \right]^{1/2} $$

$$ E_\chi \approx m_\chi + \frac{1}{2}m_\chi v_\chi^2 $$

```python
def E_chi(m_chi, v_chi):
    """
    returns: energy of DM particle in same units as m_chi [MeV]
    ********
    m_chi: DM particle mass [MeV]
    v_chi: DM particle velocity [km/s]
    """
    V_CHI = v_chi / (c/1000) # convert from [km/s] to [unitless]
    return ( m_chi + (1/2) * m_chi * V_CHI**2 ) # [MeV]


def v_chi_plus(m_N, m_chi, deltaE, v_i):
    """
    returns: [km/s] v_chi
    ********
    m_N: [MeV] nucleus mass
    m_chi: [MeV] DM mass
    deltaE: [MeV] excitation energy
    v_i: [unitless] nucleus velocity
    """
    # print(deltaE - m_chi - (1/2) * m_N * v_i**2)
    if (deltaE - m_chi - (1/2) * m_N * v_i**2) < 0:
        # print('sqrt less than 0')
        return 0
    return np.sqrt( (2 * (deltaE - m_chi - (1/2) * m_N * v_i**2) ) / (m_chi) ) * (c/1000) # [km/s]

# def norm_v_chi(s, b, l, V_ESC, sigma_chi):
#     #V_ESC = v_esc(s, b, l) # [km/s]
#     # print(V_ESC) # [km/s]
#     # sigma_chi = v_dispersion('DM', s, b, l) # [km/s]
#     # print(sigma_chi) # [km/s]
#     norm = (1 / (4 * np.pi)) * 1/( np.sqrt(np.pi/2) * sigma_chi**3 * scipy.special.erf(V_ESC / (np.sqrt(2) * sigma_chi) ) - sigma_chi**2 * V_ESC * np.exp(-V_ESC**2 / (2 * sigma_chi**2)))
#     return norm

def norm_chi(V_ESC, sigma_chi):
    N_chi = (2*np.pi * sigma_chi**2)**(-3/2) * (4 * np.pi) \
    * ( np.sqrt(np.pi / 2) * sigma_chi**3 * scipy.special.erf(V_ESC / (np.sqrt(2) * sigma_chi) ) \
       - sigma_chi**2 * V_ESC * np.exp(-V_ESC**2 / (2 * sigma_chi**2)))
    return N_chi

def Int_v_chi(nucleus, m_N, m_chi, deltaE, s, b, l, v_i):
    """
    returns: integral of DM velocity distribution [MeV^-1 km s^-1]
    **********
    deltaE: excitation energy [MeV]
    m_chi: DM mass [MeV] 
    m_N: nucleus mass [MeV]
    v_i: nucleus velocity [km/s]
    r: radius from galactic centre [kpc]
    """
    # v_i /= const.c.to('km/s') # convert from km/s to unitless
    # print(v_i)
    V_I = v_i / (c/1000) # convert from km/s to unitless
    if m_chi < 0:
        # print("m_chi < 0")
        return 0
    if (deltaE - m_chi - (1/2) * m_N * V_I**2) < 0: # process is not possible
        # print("sqrt < 0 ")
        return 0
    
    sigma_chi = v_dispersion('DM', s, b, l) # [km/s]
    v_chi = v_chi_plus(m_N, m_chi, deltaE, V_I) # [km/s]
    V_ESC = v_esc(s, b, l) # [km/s]
    norm = norm_chi(V_ESC, sigma_chi)
    # print(norm)
    
    # print('v_chi = {}'.format(v_chi))
    if v_chi > V_ESC:
        # print('v_chi > v_esc')
        return 0
    # print(v_chi)

    return ( (1 / norm) * (1 / E_chi(m_chi, v_chi)) * v_chi * np.exp(-v_chi**2 / (2 * sigma_chi**2)) ) # [MeV^-1 km/s]
```

## $v_N$ bounds

From $V_\chi^+ (v_N)$ we can find the bounds for $v_N$ that will give a nonzero answer to further narrow down the integration range and get a more accurate result.
$$ V_\chi^+ (v_i) = \left[ \frac{2(\Delta E - m_\chi - \frac{1}{2}m_N v_N^2)}{m_\chi} \right]^{1/2} $$

### $v_N$ upper bound:
$$\Delta E - m_\chi - \frac{1}{2} m_N v_N^2 >0 $$
\begin{equation}
\boxed{v_N < \left[ \frac{2(\Delta E - m_\chi)}{m_N} \right]^{1/2}}
\end{equation}


### $v_N$ lower bound:
$$ \left[ \frac{2(\Delta E - m_\chi - \frac{1}{2}m_N v_N^2)}{m_\chi} \right]^{1/2} < v_\text{esc} $$
\begin{equation}
\boxed{\left[ \frac{2(\Delta E - m_\chi - \frac{1}{2}m_\chi v_\text{esc}^2)}{m_N} \right]^{1/2} < v_N}
\end{equation}

```python
def find_v_N_bounds(m_N, m_chi, deltaE, s, b, l):
    """
    returns: [km/s] min and max bounds on v_N
    """
    V_ESC = v_esc(s, b, l) # [km/s]
    v_N_min = np.sqrt(2 * (deltaE - m_chi - (1/2) * m_chi * (V_ESC/(c/1000))**2) / m_N) * (c/1000) # [km/s]
    v_N_max = np.sqrt(2 * (deltaE - m_chi) / m_N) * (c/1000) # [km/s]
    
    if v_N_min > V_ESC:
        v_N_min = V_ESC
    if v_N_max > V_ESC:
        v_N_max = V_ESC
    if v_N_min > v_N_max:
        print('v_N bounds make no sense!')
        
    return [v_N_min, v_N_max] # [km/s]


def E_N(m_N, v_i):
    """
    returns: [MeV] non relativistic nucleus energy in same units as the nuclear mass
    ********
    m_N: [MeV] nucleus mass
    v_i: [km/s] baryon velocity
    """
    # v_i /= const.c.to('km/s') # convert from km/s to unitless
    V_I = v_i / (c/1000) # convert from km/s to unitless
    return ( m_N + (1/2) * m_N * V_I**2 ) # [MeV]


def norm_b(VBAR, V_ESC, sigma_b):
    norm = (2*np.pi * sigma_b**2)**(-3/2) * (4 * np.pi) \
    * (1/2) * sigma_b * (np.sqrt(2 * np.pi) * (VBAR**2 + sigma_b**2) * scipy.special.erf(VBAR / (np.sqrt(2) * sigma_b)) \
                        - np.sqrt(2 * np.pi)  * (VBAR**2 + sigma_b**2) * scipy.special.erf((VBAR - V_ESC) / (np.sqrt(2) * sigma_b)) \
                        + 2 * sigma_b**2 * VBAR * np.exp(- VBAR**2 / (2 * sigma_b**2)) \
                        - 2 * 2 * sigma_b**2 * (VBAR + V_ESC) * np.exp(- (VBAR - V_ESC)**2 / (2 * sigma_b**2)) )

    # print(norm)
    return norm


def Int_v_i(nucleus, m_N, m_chi, deltaE, s, b, l):
    """
    returns: integral of baryon velocity distribution
    **********
    deltaE: [MeV] excitation energy
    m_chi: [MeV] DM mass
    nucleus: scattering target
    r: [kpc] radius from galactic centre
    """
    if m_chi > deltaE: # process not possible
        return 0

    VBAR = vbar # [km/s]
    # print(VBAR)
    V_ESC = v_esc(s, b, l) # [km/s]
    # print(V_ESC)
    if VBAR>V_ESC:
        return 0
    
    sigma_b = v_dispersion(nucleus, s, b, l) # [km/s]
    # print(sigma_b)
    v_N_min = find_v_N_bounds(m_N, m_chi, deltaE, s, b, l)[0] # [km/s]
    # print(v_N_min)
    v_N_max = find_v_N_bounds(m_N, m_chi, deltaE, s, b, l)[1] # [km/s]
    # print(v_N_max)
    norm = norm_b(VBAR, V_ESC, sigma_b)
    if norm<0:
        return 0
    # print(norm)
    
    def f(v_b):
        # v_i [km/s]
        # print("v_i = {}".format(v_i))    
        integrand = ((2 * np.pi) * (1 / E_N(m_N, v_b)) * sigma_b**2 * (v_b / VBAR) \
        * (np.exp(- (v_b - VBAR)**2 / (2 * sigma_b**2) ) - np.exp(- (v_b + VBAR)**2 / (2 * sigma_b**2) )) \
        * Int_v_chi(nucleus, m_N, m_chi, deltaE, s, b, l, v_b)) 
        # [MeV^-1] * [km/s]^2 * [MeV^-1 * km/s]
        return integrand # [MeV^-2 * km^3/s^3]

    # print(scipy.integrate.quad(f, 0, v_esc))
    return (1/norm) * scipy.integrate.quad(f, v_N_min, v_N_max)[0] # [MeV^-2 km^4 s^-4]
    
```

# $dN/dE$

$$ \frac{dN^{i\chi}}{dE_\gamma} = \delta(\Delta E - E_\gamma),$$
for true gamma ray energy $E_\gamma$
$$ R_\epsilon(E-E_\gamma) \approx \frac{1}{\sqrt{2\pi\epsilon^2 E_\gamma^2}} \exp \left[ -\frac{(E-E_\gamma)^2}{2\epsilon^2 E_\gamma^2} \right]$$

\begin{align}
\int dE_\gamma ~\frac{dN^{i\chi}}{dE_\gamma} R_\epsilon(E-E_\gamma) &= \int dE_\gamma ~\delta(\Delta E - E_\gamma) ~R_\epsilon(E-E_\gamma) \\
&= R_\epsilon(E-\Delta E) \\
&= \frac{1}{\sqrt{2\pi\epsilon^2 (\Delta E)^2}} \exp \left[ -\frac{(E-\Delta E)^2}{2\epsilon^2 (\Delta E)^2} \right]
\end{align}

```python
def R_epsilon(epsilon, E, deltaE):
    """
    returns gaussian function in units [MeV^-1]
    ********
    epsilon:
    E: observed photon energy [MeV]
    E_gamma: true photon energy [MeV]
    """
    
    return (np.sqrt(2 * np.pi) * epsilon * deltaE)**(-1) * np.exp(-(E - deltaE)**2 / (2 * epsilon**2 * deltaE**2)) # [MeV^-1]
```

# line of sight integral

```python
def Int_los(nucleus, m_N, m_chi, deltaE, b, l, R_max):
    """
    returns: [cm^-5] line of sight integral
    ********
    """
    if m_chi > deltaE:
        return 0
    
    def f(s): 
        rho_i = density(nucleus, s, b, l) # [GeV/cm^3]
        # print(rho_i)
        rho_chi = density('DM', s, b, l) # [GeV/cm^3]
        # print(rho_chi)
        sigma_i = v_dispersion(nucleus, s, b, l) # km/s
        # print(sigma_i)
        sigma_chi = v_dispersion('DM', s, b, l) # km/s
        # print(sigma_chi)
        # print(Int_v_i(nucleus, m_N, m_chi, deltaE, s, b, l))
        
        return rho_i * rho_chi * ( (2*np.pi) * sigma_i * sigma_chi )**(-3) * Int_v_i(nucleus, m_N, m_chi, deltaE, s, b, l)
        # [GeV/cm^3] * [GeV/cm^3] * [km/s]^-3 * [km/s]^-3 * [MeV^-2 km^4 s^-4] = [GeV^2 cm^-6 km^-2 s^2 MeV^-2]
    # print(s_max(b, l, R_max))
    # result = [GeV^2 cm^-6 km^-2 s^2 MeV^-2]
    MeV_per_GeV = 1000
    integral = quad(f, 0, s_max(b, l, R_max))[0]
    
    return (integral * MeV_per_GeV**2 * (c/1000)**2 * cm_per_kpc) # [cm^-5]
    # integrating over kpc gives a factor of kpc
    
```

```python
def flux_w_conv(g_chi, g_A, nucleus, m_chi, b, l, R_max, epsilon, E):
    """
    returns: differential flux in [cm^-2 s^-1 (sr^-1) MeV^-1]
    **********
    g_chi: [MeV^-1]
    g_A: axial form factor [unitless]

    E: observed photon energy [MeV]
    """
    nuc_info = {'He4': [3758.26, 0.0], 'C12': [11274.78, 0.0], 'N14': [13153.91, 0.0], 'O16': [15033.04, 0.0]}
    # datafile = '{}_{}_{}.txt'.format(nucleus, nuc_info[nucleus][0], nuc_info[nucleus][1])
    # df = pd.read_csv(datafile, sep='\t', names=['E_gamma [MeV]', 'GT strength'], skiprows=1)
    
    flux_arr = []
    flux_tot = 0
    #nucDict = {'He4': 'he4', 'C12': 'c12', 'N14': 'n14', 'O16': 'o16', 'Fe57': 'fe57'}

    m_N = nuc_info[nucleus][0]
    J_N = nuc_info[nucleus][1]
    dEs = dEs_dict[nucleus]
    GTs = GTs_dict[nucleus]
    nucl_exc = zip(dEs, GTs)
    GT_sum = np.sum(GTs)

    # print(dEs)
    for dE, GT in nucl_exc:
        # print(dE)
        Int = Int_los(nucleus, m_N, m_chi, dE, b, l, R_max) # [cm^-5]
        print("los integral = {}".format(Int))
        dNdgamma = GT / GT_sum # [unitless]
        R = 1 #R_epsilon(epsilon, E, dE) # [MeV^-1]
        flux = ((1/24 * g_chi**2 * g_A**2 / (2*J_N + 1)) * (m_N + dE)/m_N * Int * dNdgamma * R * (100*c) * (cmMeV)**2)
        flux_arr.append(flux) # [MeV^-2 * cm^-5 * MeV^-1 * cm/s * cm^2 * MeV^2] = [cm^-2 s^-1 (sr^-1) MeV^-1]
        flux_tot += flux
    
    return(dEs, np.array(flux_arr), flux_tot) # [cm^-2 s^-1 MeV^-1]
```

```python
def CO_flux_fx(folder, g_chi, g_A, m_chi, b, l, R_max, epsilon):
    """
    computes CO flux - takes a few mins!
    ********
    epsilon: decimal from 0 to 1
    """
    energy = np.logspace(np.log10(m_chi/10), np.log10(m_chi*10), 51)
    CO_flux = []
    for E in energy:
        O_flux = flux_w_conv(g_chi, g_A, 'O16', m_chi, b, l, R_max, epsilon, E)[2]
        C_flux = flux_w_conv(g_chi, g_A, 'C12', m_chi, b, l, R_max, epsilon, E)[2]
        CO_flux.append(O_flux + C_flux)

    dat = {'Photon energy [MeV]': energy, 'CO flux [cm^-2 s^-1 MeV^-1]': CO_flux}
    #, 'g_chi': 1, 'g_A': 1, 'b (deg)': b_crval, 'l [deg]': l_crval, 'R_max [kpc]': 10*R_odot}
    df = pd.DataFrame(dat)
    df.to_csv('/Users/arm5765/Desktop/git/DM-nucleus-inelastic-ID/fluxData/{}/mchi_{}MeV_gA_{}_gchi_{}_b_{}deg_l_{}deg_Rmax_{}kpc_epsilon_{}newunits.txt'.format(folder, m_chi, g_A, g_chi, b, l, R_max, 100*epsilon), sep='\t', index=False)
    # df.to_csv('mchi_{}MeV_gA_{}_gchi_{}_b_{}deg_l_{}deg_Rmax_{}kpc_epsilon_{}newunits.txt'.format(m_chi, g_A, g_chi, b, l, R_max, 100*epsilon), sep='\t', index=False)
    return CO_flux
```
